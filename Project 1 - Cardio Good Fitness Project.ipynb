##### Boxplots
Provides a way of statistically representing the distribution of the data using five main dimensions - medium, minimum, maximum, first and third quartiles. 

Boxplots summarize the distribution of each attribute, drawing a line for the median (middle value) and a box around the 25th and 75th percentiles (the middle 50% of the data). The whiskers give an idea of the spread of the data and dots outside of the whiskers show candidate outlier values.

#### Multivariate Analysis

```python
sns.pairplot(dataframe);
```

With the pairwise plot we can visualise a lot of the variables in the dataset as it provides a quick visual image of the dataset before more exploratory analysis.
As we can see, there is a stong corelation between income and age, Income and Miles, Usage and Mile and Fitness and Miles.

```python
sns.lmplot(y="Income", x = 'Education', data = dataframe);
```

```python
sns.lmplot(y="Miles", x = 'Income', data = dataframe);
```

There is a positive corelation between income and age as well as between income and miles. The corelation is not the strongest at 0.5 however there is a relationship between income and miles. The relationship is stronger for income and age at 0.7 which is closer to perfect corelation of 1 than 0.5

```python
plt.figure(figsize = (10,5))
sns.heatmap(dataframe.corr('spearman'), annot = True, linewidths = 0.5, fmt = '.1f', center = 1);
```

The result of the heat map shows that there is a strong positive corelation between miles and usage, miles and fitness, usage and fitness and education and income. The corelation is not as strong in education and income as it is 0.6 compared to the other which are 0.7 and 0.8
